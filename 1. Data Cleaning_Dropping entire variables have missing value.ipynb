```python
#Getting the name of the columns have object values
object_clm = data3_drop_rows.select_dtypes(include=['object']).columns
object_clm
```

```python
#Checking the orginal data and clean data
def cat_var_distr(var):
  return pd.concat([data[var].value_counts() / data.shape[0] * 100,
                    data3_drop_rows[var].value_counts() / data3_drop_rows.shape[0] * 100,
                    (data[var].value_counts() / data.shape[0] * 100) - (data3_drop_rows[var].value_counts() / data3_drop_rows.shape[0] * 100)],
                    axis = 1, keys = [var + '_org', var + '_clean', 'Balance'])

for i in object_clm:
  display(cat_var_distr(i))
  display('*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^')
```
