# One Hot Encoding

```python
#Importing library
from sklearn.preprocessing import OneHotEncoder
```

```python
#Calling one hot encoder function
one_hot_encode = OneHotEncoder(sparse=False)
```

```python
#Applying the one hot encoder on data
one_hot_encode_data = one_hot_encode.fit_transform(tips_data[['sex','smoker','day','time']])
```

```python
#View the data
one_hot_encode_data
#This data are in the form of numpy array
```

```python
#Extracting column names
tips_data_dummy.columns
```

```python
#Converting numpy array data into data frame
one_hot_encode_dataframe = pd.DataFrame(one_hot_encode_data, columns=['sex_Female', 'sex_Male', 'smoker_No',
       'smoker_Yes', 'day_Fri', 'day_Sat', 'day_Sun', 'day_Thur',
       'time_Dinner', 'time_Lunch'])
```
