# MISSING VALUES ANALYSIS

```python
#Getting missing values count with variables list
data.isnull().sum()
```

```python
#Getting missing values count with variables list
data.isnull().sum().sum()
```

```python
#Getting missing values count with variables list except zero
data.isnull().sum()[data.isnull().sum()>0]
```

```python
#Getting proportinal missing values with variables list
missing_value_per = data.isnull().sum()[data.isnull().sum()>0]/data.shape[0] * 100
missing_value_per
```

```python
#Extracting variables name have missing values greater than 20%
missing_value_clm_gre_20 = missing_value_per[missing_value_per>20].keys()
missing_value_clm_gre_20
```

# VARIABLES DROPPING - MISSING VALUES GREATER THAN 20% 

```python
#Creating data set with dropping missing values greater than 20%
data2_drop_clm = data.drop(columns = missing_value_clm_gre_20)
```

```python
#Checking the shape of data
data2_drop_clm.shape
```

# NUMERICAL DATA ANALYSIS

```python
#Creating data set having only numerical variables
data3_num = data2_drop_clm.select_dtypes(include = ['int64','float64'])
```

```python
#View the top five rows
data3_num.head()
```

```python
#Checking the shape of the data
data3_num.shape
```

# MISSING VALUES ANALYSIS OF NUMERICAL DATA

```python
#Heatmap plotting on numerical data
plt.figure(figsize = (16,9))
sns.heatmap(data3_num.isnull())
```

```python
#Extracting numerical data having missing values
data3_num[data3_num.isnull().any(axis = 1)]
```

```python
#Checking the shape of the data
data3_num[data3_num.isnull().any(axis = 1)].shape
```

```python
#Getting list variable with missing value counts in the data
data3_num.isnull().sum()
```

```python
#Getting sum of missing value counts in the data
data3_num.isnull().sum().sum()
```

```python
#Getting list variable with missing value counts in the data except zero
missing_num_var = [var for var in data3_num.columns if data3_num[var].isnull().sum()>0]
missing_num_var
```

```python
#Plotting distribution plot
plt.figure(figsize = (10,10));
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(data3_num[var], bins = 20, kde_kws = {'linewidth':5, 'color':'#DC143C'})
```

# FILLING MISSING VALUES BY MEAN IN NUMERICAL DATA

```python
#Filling missing values by mean
data4_num_mean = data3_num.fillna(data3_num.mean())
```

# MISSING VALUES ANALYSIS OF NUMERICAL DATA

```python
#Getting total missing values counts
data4_num_mean.isnull().sum().sum()
```

```python
#Checking missing values in numerical data
data4_num_mean.isnull().sum()
```

# ANALYSIS OF NUMERICAL DATA BEFORE AND AFTER MEAN IMPUTATION THROUGH DITRIBUTION PLOT

```python
#Creating distribution plot of 
plt.figure(figsize =(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(data3_num[var], bins = 20, kde_kws = {'linewidth':5, 'color':'red'}, label = 'Original')
    sns.distplot(data4_num_mean[var], bins = 20, kde_kws = {'linewidth':5, 'color':'green'}, label = 'Mean')
    plt.legend()
```

# FILLING MISSING VALUES BY MEDIAN IN NUMERICAL DATA

```python
# Filling missing values by median in numerical data
data5_num_median = data3_num.fillna(data3_num.median())
```

```python
#Checking missing values in numerical data
data5_num_median.isnull().sum().sum()
```

# ANALYSIS OF NUMERICAL DATA, NUMERICAL DATA AFTER MEAN IMPUTATION & NUMERICAL DATA AFTER MEDIAN IMPUTATION   THROUGH DITRIBUTION PLOT

```python
#Creating ditribution plot of numerical data, numerical data after mean imputation and numerical data adter median imputation
plt.figure(figsize =(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(data3_num[var], bins = 20,hist = False, kde_kws = {'linewidth':5, 'color':'red'}, label = 'Original')
    sns.distplot(data4_num_mean[var], bins = 20,hist = False, kde_kws = {'linewidth':5, 'color':'green'}, label = 'Mean')
    sns.distplot(data5_num_median[var], bins = 20,hist = False,kde_kws = {'linewidth':5, 'color':'purple'}, label = 'Median'  )
    plt.legend()
```

# ANALYSIS - COMBINING NUMERICAL DATA, NUMERICAL DATA AFTER MEAN IMPUTATION AND NUMERICAL DATA AFTER MEDIAN IMPUTATION

```python
#Combining numerical data, numerical data after mean imputation and numerical data after median imputation
data_concat = pd.concat([data3_num[missing_num_var], 
                         data4_num_mean[missing_num_var], 
                         data5_num_median[missing_num_var]], axis = 1)
```
