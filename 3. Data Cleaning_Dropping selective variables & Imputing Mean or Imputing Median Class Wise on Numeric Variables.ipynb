# SINGLE MISSING VALUE VARIABLE MEAN IMPUTATION CLASS WISE FOR ONE CLASS

```python
#Replacing missing values in the varaible LotConfig by mean in sub class Inside 
data[data.loc[:,'LotConfig']=='Inside']['LotFrontage'].replace(np.nan, data[data.loc[:,'LotConfig']=='Inside']['LotFrontage'].mean())
```

# SINGLE MISSING VALUE VARIABLE MEAN IMPUTATION CLASS WISE FOR ONE CLASS THROUGH FOR LOOP FUNCTION

```python
#Creating a copy of the data
data_copy = data.copy()
data_copy.head()
```

```python
#Replacing NA value in the varaible lotFrontage by mean of LotFrontage as per the category of LotConfig 
for var_class in data['LotConfig'].unique():
    data_copy.update(data[data.loc[:,'LotConfig']=='Inside']['LotFrontage']
                     .replace(np.nan, data[data.loc[:,'LotConfig']=='Inside']['LotFrontage'].mean()))
```

# ANALYSIS OF MISSING VALUES AFTER ONE MISSING VALUE VARIABLE MEAN IMPUTATION

```python
#Checking the count of missing values
data_copy.isnull().sum()
```

# SINGLE MISSING VALUE VARIABLE MEAN IMPUTATION CLASS WISE FOR MULTIPLE CLASS

```python
#Mean imputation in the missing values variable LotConfig's all the sub class
data_copy = data.copy()
for var_class in data['LotConfig'].unique():
    data_copy.update(data[data.loc[:,'LotConfig']==var_class]['LotFrontage'].replace(np.nan, data[data.loc[:,'LotConfig']==var_class]['LotFrontage'].mean()))
```

```python
#Checking missing values
data_copy.isnull().sum()
```

# MULTIPLE MISSING VALUE VARIABLES MEAN IMPUTATION CLASS WISE FOR MULTIPLE CLASS(IMPLEMENTATION UNSUCCESSFULL DUE CATEGORICAL CLASSES HAVE MISSING VALUES)

```python
#Mean imputation in all the missing value variables sub classes
data_copy = data.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig','MasVnrType','GarageType']
for cat_var, num_var_miss in zip(cat_vars,num_vars_miss):
    for var_class in data[cat_var].unique():
        data_copy.update(data[data.loc[:,cat_var] == var_class][num_var_miss].replace(np.nan,data[data.loc[:,cat_var] == var_class][num_var_miss].mean()))
```

```python
#Checking missing value
data_copy[num_vars_miss].isnull().sum()
```

```python
#Checking the data
data_copy[data_copy[['MasVnrType']].isnull().any(axis =1)]
```

```python
##Checking the data
data_copy[data_copy[['GarageType']].isnull().any(axis =1)]
```

# MULTIPLE MISSING VALUE VARIABLES MEAN IMPUTATION CLASS WISE FOR MULTIPLE CLASS

```python
#Mean imputation in all the missing value variables sub classes
data_copy = data.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig','Exterior2nd','KitchenQual']
for cat_var, num_var_miss in zip(cat_vars,num_vars_miss):
    for var_class in data[cat_var].unique():
        data_copy.update(data[data.loc[:,cat_var] == var_class][num_var_miss].replace(np.nan,data[data.loc[:,cat_var] == var_class][num_var_miss].mean()))
```

```python
#Checking missing values
data_copy[num_vars_miss].isnull().sum()
```

# DISTRIBUTION PLOT BEFORE & AFTER MEAN IMPUTATION

```python
#Dstribution plot on before & after mean imputation
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(num_vars_miss):
    plt.subplot(2,2,i+1)
    sns.distplot(data[var], bins=20, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(data_copy[var], bins=20, kde_kws={'linewidth':5, 'color':'green'},label="Mean",)
    plt.legend()
```

# MULTIPLE MISSING VALUE VARIABLES MEDIAN IMPUTATION CLASS WISE FOR MULTIPLE CLASS

```python
data_copy_median = data.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig','Exterior2nd','KitchenQual']
for cat_var, num_var_miss in zip(cat_vars,num_vars_miss):
    for var_class in data[cat_var].unique():
        data_copy_median.update(data[data.loc[:,cat_var] == var_class][num_var_miss].replace(np.nan,data[data.loc[:,cat_var] == var_class][num_var_miss].median()))
```

```python
#Checking missing values
data_copy_median[num_vars_miss].isnull().sum()
```

# DISTRIBUTION PLOT BEFORE & AFTER MEAN IMPUTATION & MEDIAN IMPUTATION

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(num_vars_miss):
    plt.subplot(2,2,i+1)
    sns.distplot(data[var], bins=20, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(data_copy[var], bins=20, kde_kws={'linewidth':5, 'color':'green'},label="Mean",)
    sns.distplot(data_copy_median[var], bins=20, kde_kws={'linewidth':3, 'color':'k'},label="Median",)
    plt.legend()
```

# BOXPLOT BEFORE & AFTER MEAN IMPUTATION & MEDIAN IMPUTATION

```python
for i, var in enumerate(num_vars_miss):
    plt.figure(figsize = (10,10))
    plt.subplot(3,1,1)
    sns.boxplot(data[var])
    plt.subplot(3,1,2)
    sns.boxplot(data_copy[var])
    plt.subplot(3,1,3)
    sns.boxplot(data_copy_median[var])
```
